```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('MP_2022.xlsx')
```

```python
df = df.drop('ID',axis=1)
df = df.drop('SEC',axis=1)
df = df.drop('GEN',axis=1)
df = df.drop('FATHERS_NAME',axis=1)

df['ID'] = df.index + 1

ID_COLUMN = df.pop('ID')
df.insert(0,'ID',ID_COLUMN)
```

```python
df.fillna(0,inplace=True)
```

```python
id = df['ID']
df['TOTAL'] = df['BNGA']+df['ENGB']+df['PSC']+df['LSC']+df['MATH']+df['HIST']+df['GEO']
df['PERCENT'] = (df['TOTAL']/700)*100
```

```python
bengali = df['BNGA']
english = df['ENGB']
physicalScience = df['PSC']
lifeScience = df['LSC']
math = df['MATH']
history = df['HIST']
geography = df['GEO']
totalPercentage = df['PERCENT']
```

```python
plt.xlabel('Serial Number')
plt.ylabel('Marks in Bengali')
plt.vlines(x = 147, ymin = 0, ymax = 100,
           colors = 'red',
           label = 'vline_multiple - full height')
plt.scatter(id,bengali)
```
