```python
ine=pd.read_csv("../input/income-inequality/combined_final_last_10_years.csv")
ine.head(10)
```

### The gini index is a measure of statistical dispersion intended to represent the income inequality or the wealth inequality within a nation or a social group. 

```python
ine.shape
```

```python
ine.isnull().sum() # No null values :)
```

```python
data1=ine.loc[ine["country"].isin(["United Kingdom","Japan","Australia","Germany","Norway","Finland","New Zealand","Luxembourg","France","Canada"])]
data2=ine.loc[ine["country"].isin(["India","China","Brazil","Russia","Bangladesh","Pakistan","Mexico","Chile","Sri Lanka","Indonesia","South Africa"])]
```

## EDA

```python

ax=sns.catplot(y='gini_index',x='year',hue='country',kind='point',height=6,aspect=2,data=data1)
ax.fig.suptitle('First World')
```

```python
ax=sns.catplot(y='gini_index',x='year',hue='country',kind='point',height=6,aspect=2,data=data2)
ax.fig.suptitle('Second World')
```

### Most of the second world countries are either flat or declining in terms of income inequality, which is a good thing.

### Comparison between democracy quality and income inequality.

```python
sns.catplot(y='gini_index',x='demox_eiu',hue='country',height=8,aspect=2,data=data1)
```

## Conclusions
#### Most of the first world countries have relatively low inequalities than second world countries.
#### There is a good correlation between good democracy and income inequality, countries with good governance have low inequality.
