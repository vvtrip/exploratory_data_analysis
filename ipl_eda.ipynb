```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/ipl_matches.csv").set_index('id')
df.head()
```

```python
plt.figure(figsize=(14, 6))
df.player_of_match.value_counts()[:15].plot(kind="bar")
plt.title("Top 15 players with most number of player of the match")
```

```python
plt.figure(figsize=(14, 6))
df.season.value_counts().plot(kind='bar')
plt.title('Seasons with most number of matches')
```

```python
plt.figure(figsize=(14, 6))
df.venue.value_counts().plot(kind='bar')
plt.title('Most used venues')
```

```python
plt.figure(figsize=(14,6))
df.groupby('season').toss_decision.value_counts().plot(kind='bar')
plt.title('Toss decision by season')
```

```python
plt.figure(figsize=(14,6))
df.winner.value_counts().plot(kind='bar')
plt.title('Most wins by Team')
```

```python
plt.figure(figsize=(14,6))
df.toss_winner.value_counts().plot(kind='bar')
plt.title('Most Toss wins by Team')
```
